```python
import pandas as pd
import numpy as np

# ######### 2025.02.02 22:00:00 ################################## ADATDOKI ###
# csv_df adatok feldolgozása függvényként: időrendi sorrendben, max 22 sor, napszakonként egy sorba rendezve
# #########1#########2#########3#########4#########5#########6#########7#########8

def feldolgoz_vernyomas_adatok(csv_df: pd.DataFrame, kezdo_datum: str, max_sor: int = 22):
    """
    Vérnyomásadatok feldolgozása a megadott kezdő dátumtól, max. 22 sorig.

    Bemenet:
    - csv_df (pd.DataFrame): Az összes vérnyomásmérési adatot tartalmazó DataFrame.
    - kezdo_datum (str): A kezdő dátum (ÉÉÉÉ-HH-NN formátumban).
    - max_sor (int, alapértelmezés: 22): A maximális sorok száma.

    Kimenet:
    - oldal_df (pd.DataFrame): A formázott adatok.
    - utolso_datum (str): Az utolsó beillesztett dátum.
    - hibakod (str vagy None): Hiba esetén egy hibakód, egyébként None.
    """

    # 🔹 Fejléc beállítása
    columns = ["sor", "Dátum", "reg_sys", "reg_dia", "reg_pul",
               "nk_sys", "nk_dia", "nk_pul",
               "est_sys", "est_dia", "est_pul"]

    # Üres oldal_df létrehozása
    oldal_df = pd.DataFrame(columns=columns)

    # 🔹 Kezdő dátum ellenőrzése és konvertálása
    try:
        validated_date = pd.to_datetime(kezdo_datum, format="%Y-%m-%d").date()
    except ValueError:
        return None, None, "HIBA_ROSSZ_DATUM"

    # 🔹 Ha csv_df nem létezik vagy üres, hibát adunk vissza
    if csv_df.empty:
        return None, None, "HIBA_CSV_ÜRES"

    # 🔹 Ellenőrizzük, hogy a szükséges oszlopok léteznek-e
    required_columns = {"Dátum", "Idő", "Sys", "Dia", "Pulzus"}
    if not required_columns.issubset(csv_df.columns):
        return None, None, "HIBA_HIANYZO_OSZLOPOK"

    # 🔹 A csv_df["Dátum"] oszlopot datetime formátumra alakítjuk
    csv_df["Dátum"] = pd.to_datetime(csv_df["Dátum"], errors="coerce").dt.date

    # 🔹 Az `Idő` oszlopot `H:M:S` formátumból órává alakítjuk
    csv_df["Óra"] = pd.to_datetime(csv_df["Idő"], format="%H:%M", errors="coerce").dt.hour

    # 🔹 A csv_df-et időrend szerint rendezzük (Dátum, majd Idő növekvő sorrendben)
    csv_df = csv_df.sort_values(by=["Dátum", "Óra"], ascending=[True, True])

    # 🔹 Kezdő index és sor sorszámozás
    sor_index = 1

    # 🔹 Az oldal_df feltöltése
    for _, meres in csv_df.iterrows():
        datum = meres["Dátum"]
        ora = meres["Óra"]

        if datum < validated_date:
            continue  # Ha a dátum korábbi, mint a megadott kezdő dátum, kihagyjuk

        # Megkeressük az utolsó olyan sort, ahol az adott dátum szerepel, és van még szabad hely
        lehetoseg = oldal_df[(oldal_df["Dátum"] == datum) & (
            (0 <= ora < 9 and oldal_df["reg_sys"].isna()) |
            (9 <= ora < 18 and oldal_df["nk_sys"].isna()) |
            (ora >= 18 and oldal_df["est_sys"].isna())
        )]

        if not lehetoseg.empty:
            # Van olyan meglévő sor, ahová még beilleszthetjük
            sor_index_beilleszt = lehetoseg.index[-1]
        else:
            # Ha nincs szabad hely, új sort kell nyitni
            if len(oldal_df) >= max_sor:  # Ha már max_sor van, akkor leállunk
                break
            sor_index_beilleszt = len(oldal_df)
            oldal_df.loc[sor_index_beilleszt] = [sor_index, datum] + [np.nan] * (len(columns) - 2)
            sor_index += 1

        # 🔹 A megfelelő oszlopba írjuk az értékeket
        if 0 <= ora < 9:
            oldal_df.loc[sor_index_beilleszt, ["reg_sys", "reg_dia", "reg_pul"]] = meres[["Sys", "Dia", "Pulzus"]].values
        elif 9 <= ora < 18:
            oldal_df.loc[sor_index_beilleszt, ["nk_sys", "nk_dia", "nk_pul"]] = meres[["Sys", "Dia", "Pulzus"]].values
        elif ora >= 18:
            oldal_df.loc[sor_index_beilleszt, ["est_sys", "est_dia", "est_pul"]] = meres[["Sys", "Dia", "Pulzus"]].values

    # 🔹 Meghatározzuk az utolsó dátumot
    if not oldal_df.empty:
        utolso_datum = str(oldal_df["Dátum"].iloc[-1])
    else:
        utolso_datum = None

    # 🔹 Sikeres végrehajtás esetén hibakód nincs
    return oldal_df, utolso_datum, None


# 📌 Függvény meghívása
kezdo_datum = "2024-10-15"
oldal_df, utolso_datum, hibakod = feldolgoz_vernyomas_adatok(csv_df, kezdo_datum, max_sor=22*18)
print(utolso_datum)
#print(hibakód)
di(oldal_df)
```

```python
from openpyxl import load_workbook
import pandas as pd
import re  # Reguláris kifejezések az oszlop és sor szétválasztására

# ######### 2025.02.02 22:45:00 ################################## ADATDOKI ###
# oldal_df adatainak beírása 18 db Excel blokkba 22 soronként, az eredeti formátum megtartásával
# #########1#########2#########3#########4#########5#########6#########7#########8

def szetvalaszt_oszlop_sor(cella: str):
    """
    Egy Excel cella címét (pl. 'AB4') szétválasztja az oszlop és sor részre.

    Bemenet:
    - cella (str): Az Excel cella címe (pl. 'F4', 'AB4')

    Kimenet:
    - oszlop_betu (str): Az oszlop betűje (pl. 'F', 'AB')
    - sor_szam (int): A sor száma (pl. 4)
    """
    match = re.match(r"([A-Z]+)([0-9]+)", cella)
    if not match:
        raise ValueError(f"Hibás cella formátum: {cella}")

    return match.group(1), int(match.group(2))


def irj_excel_blokkokba(file_path: str, oldal_df: pd.DataFrame):
    """
    Az oldal_df tartalmát 22 soros blokkokra osztva beírja az Excel fájl megfelelő tartományaira.

    Bemenet:
    - file_path (str): Az Excel fájl elérési útja.
    - oldal_df (pd.DataFrame): A beírandó adatok (táblázatos formátum).

    Kimenet:
    - Sikeres beírás esetén kiírja: "Az adatok sikeresen beírva az Excel fájlba."
    - Hiba esetén kiírja a hibát.
    """

    # 📌 Excel fájl betöltése
    try:
        wb = load_workbook(file_path)
        ws = wb["napló_c"]  # A cél munkalap
    except Exception as e:
        print(f"Hiba az Excel fájl megnyitásakor: {e}")
        return

    # 📌 18 blokk kezdőcelláinak listája
    blokk_kezdopontok = [
    # Első sor
    "E4", "P4", "AA4",
    # Második sor
    "E30", "P30", "AA30",
    # Harmadik sor
    "E56", "P56", "AA56",
    # Negyedik sor
    "E82", "P82", "AA82",
    # Ötödik sor
    "E108", "P108", "AA108",
    # Hatodik sor
    "E134", "P134", "AA134"
]


    # 📌 Az oldal_df-et 22 soros blokkokra osztjuk
    max_sor = 22
    blokkok_szama = len(blokk_kezdopontok)

    for blokk_idx, blokk_start in enumerate(blokk_kezdopontok):
        start_sor = blokk_idx * max_sor
        end_sor = start_sor + max_sor

        if start_sor >= len(oldal_df):
            break  # 🔹 Kilépés, ha nincs több adat

        blokk_adatok = oldal_df.iloc[start_sor:end_sor]

        if blokk_adatok.empty:
            continue  # 🔹 Üres blokk kihagyása

        start_col_letter, start_row = szetvalaszt_oszlop_sor(blokk_start)
        start_col_index = sum((ord(char) - ord("A") + 1) * (26 ** idx) for idx, char in enumerate(reversed(start_col_letter)))

        for i, row in enumerate(blokk_adatok.itertuples(index=False), start=start_row):
            for j, value in enumerate(row, start=start_col_index):
                if ws.cell(row=i, column=j).value is None:  # 🔹 Csak üres cellába írunk
                    ws.cell(row=i, column=j, value=value)

        print(f"✅ Blokk {blokk_idx + 1}/{blokkok_szama} beírva ({start_sor}-{end_sor} sor)")

    try:
        wb.save(file_path)
        print("✔ Az adatok sikeresen beírva az Excel fájlba.")
    except Exception as e:
        print(f"Hiba az Excel mentésekor: {e}")

excel_utvonal = "/content/drive/MyDrive/doku/TZ/orvosi/vérnyomás250210.xlsx"
irj_excel_blokkokba(excel_utvonal, oldal_df)
```

```python
def shiftelj_balra_oszlop(oszlop_betu):

    # Az oszlopbetűt számmá alakítjuk ('A' = 1, 'B' = 2, ..., 'Z' = 26, 'AA' = 27, stb.)
    oszlop_szam = 0
    for char in oszlop_betu:
        oszlop_szam = oszlop_szam * 26 + (ord(char) - ord("A") + 1)

    # Egy oszloppal balra shifteljük
    oszlop_szam -= 1
    if oszlop_szam < 1:
        raise ValueError(f"Nem lehet az 'A' oszlopnál balrább shiftelni: {oszlop_betu}")

    # A számból visszaalakítjuk oszlopbetűvé
    shifted_betu = ""
    while oszlop_szam > 0:
        oszlop_szam -= 1
        shifted_betu = chr(oszlop_szam % 26 + ord("A")) + shifted_betu
        oszlop_szam //= 26

    return shifted_betu
    eredeti_kezdopontok = [
        "F4", "Q4", "AB4", "F30", "Q30", "AB30",
        "F56", "Q56", "AB56", "F82", "Q82", "AB82",
        "F108", "Q108", "AB108", "F134", "Q134", "AB134"
    ]

blokk_kezdopontok = [shiftelj_balra_oszlop(c[:re.search(r"\d", c).start()]) + c[re.search(r"\d", c).start():] for c in eredeti_kezdopontok]

# 📌 Az oldal_df-et 22 soros blokkokra osztjuk
print(blokk_kezdopontok )
```
